```python
def identity_block(X, filters):
    F1, F2 = filters
    skip = X
    
    X = Conv2D(filters = F1, kernel_size = (3,3), strides = (1,1), padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = Conv2D(filters = F2, kernel_size = (3,3), strides = (1,1), padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = Add()([X, skip])
    X = Activation('relu')(X)
    
    return X
```

```python
def convolutional_block(X, filters, stride = 2):
    F1, F2 = filters
    skip = X
    
    X = Conv2D(filters = F1, kernel_size = (3,3), strides = (1,1), padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    X = Conv2D(filters = F2, kernel_size = (3,3), strides = (1,1), padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    
    skip = tf.keras.layers.Conv2D(F1, (1,1), strides = (1,1))(skip)
    X = BatchNormalization(axis = 3)(X)

    X = Add()([X, skip])
    X = Activation('relu')(X)
    
    return X
```

```python
def ResNet18(input_shape = (200, 200, 3), classes = 2):
    
    
    X_input = Input(input_shape)
    X = X_input
    X = ZeroPadding2D((3, 3))(X)
    X = Conv2D(filters = 64, kernel_size = (7,7), strides = (2,2), padding = 'same')(X)
    X = BatchNormalization(axis = 3)(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((3,3), strides = (2,2))(X)

    
   
    X = convolutional_block(X, filters = [64,64], stride = 2)
    X = identity_block(X, filters = [64,64])
    
    X = convolutional_block(X, filters = [128,128], stride = 2)
    X = identity_block(X, filters = [128,128])
    
    X = convolutional_block(X, filters = [256,256], stride = 2)
    X = identity_block(X, filters = [256,256])
    
    X = convolutional_block(X, filters = [512,512], stride = 2)
    X = identity_block(X, filters = [512,512])
    
    X = AveragePooling2D((2,2), padding = 'same')(X)
    
    X = Flatten()(X)
    X = BatchNormalization()(X)
#     X = Dropout(0.5)(X)

    X = Dense(256, activation = 'relu')(X)
    X = Dense(1, activation = 'softmax')(X)
    
    model = Model(inputs = X_input, outputs = X, name = 'ResNet18')
    
    return model
    
```

```python
model = ResNet18(input_shape = (200,200,3), classes = 2)
model.compile(loss = 'binary_crossentropy', 
             optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
             metrics = ['accuracy'])
model.summary()
# callback = ReduceLROnPlateau(monitor='val_loss', patience = 2, cooldown=0, verbose=1, factor=0.6, min_lr=0.000001)
```

```python
history = model.fit(training_dataset, validation_data = validation_dataset, epochs = 5, steps_per_epoch = 2) 
```
